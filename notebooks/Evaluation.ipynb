```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
outcome = pd.read_excel("outcome/Acc_ada.xlsx", header=0)
print(outcome.head())
```

```python
outcome_epoch_1 = outcome[outcome.epoch == 1]
outcome_epoch_5 = outcome[outcome.epoch == 5]
print(outcome_epoch_1)
print(outcome_epoch_5)

outcome_epoch_1.plot(
    x='Classes',
    y=['Acc', 'F1/macro', 'F1/micro'],
    xlabel='Class Num',
    ylabel='Accuracy',
    grid=True,
    marker='.',
    title='Acc, Macro and Micro in different number of classes after 1 epoch'
)
outcome_epoch_5.plot(
    x='Classes',
    y=['Acc', 'F1/macro', 'F1/micro'],
    xlabel='Class Num',
    ylabel='Accuracy',
    grid=True,
    marker='.',
    title='Acc, Macro and Micro in different number of classes after 5 epoch'
)
```

```python
fname = 'outcome/5classes_acc_epoch40.txt'
s1 = {}
with open(fname,'r+',encoding='utf-8') as f:
    for j, line in enumerate(f.readlines()):
        if j == 0:
            s1['acc_list'] = line[:-1].split(', ')
            for i, num in enumerate(s1['acc_list']):
                s1['acc_list'][i] = float(num)
        elif j == 1:
            s1['micro_list'] = line[:-1].split(', ')
            for i, num in enumerate(s1['micro_list']):
                s1['micro_list'][i] = float(num)
        elif j == 2:
            s1['macro_list'] = line[:-1].split(', ')
            for i, num in enumerate(s1['macro_list']):
                s1['macro_list'][i] = float(num)
    #print(s1)
```

```python
fname2 = 'outcome/10classes_acc_epoch40.txt'
s2 = {}
with open(fname2,'r+',encoding='utf-8') as f:
    for j, line in enumerate(f.readlines()):
        if j == 0:
            s2['acc_list'] = line[:-1].split(', ')
            for i, num in enumerate(s2['acc_list']):
                s2['acc_list'][i] = float(num)
        elif j == 1:
            s2['micro_list'] = line[:-1].split(', ')
            for i, num in enumerate(s2['micro_list']):
                s2['micro_list'][i] = float(num)
        elif j == 2:
            s2['macro_list'] = line[:-1].split(', ')
            for i, num in enumerate(s2['macro_list']):
                s2['macro_list'][i] = float(num)
    #print(s2)
```

```python
fname3 = 'outcome/2classes_acc_epoch40.txt'
s3 = {}
with open(fname3,'r+',encoding='utf-8') as f:
    for j, line in enumerate(f.readlines()):
        if j == 0:
            s3['acc_list'] = line[:-1].split(', ')
            for i, num in enumerate(s3['acc_list']):
                s3['acc_list'][i] = float(num)
        elif j == 1:
            s3['micro_list'] = line[:-1].split(', ')
            for i, num in enumerate(s3['micro_list']):
                s3['micro_list'][i] = float(num)
        elif j == 2:
            s3['macro_list'] = line[:-1].split(', ')
            for i, num in enumerate(s3['macro_list']):
                s3['macro_list'][i] = float(num)
    #print(s3)
```

```python
epochs = range(0,40)
epoch_acc_5class = pd.DataFrame.from_dict(s1, orient="columns")
epoch_acc_5class['epochs'] = epochs
print(epoch_acc_5class.head())
epoch_acc_10class = pd.DataFrame.from_dict(s2, orient="columns")
epoch_acc_10class['epochs'] = epochs
print(epoch_acc_10class.head())
epoch_acc_2class = pd.DataFrame.from_dict(s3, orient="columns")
epoch_acc_2class['epochs'] = epochs
print(epoch_acc_2class.head())
```

```python
epoch_acc_5class.plot(
    x='epochs',
    y=['acc_list', 'micro_list', 'macro_list'],
    xlabel='Epochs',
    ylabel='Accuracy',
    grid=True,
    marker='.',
    title='Accuracy, Micro, macro with epochs for 5-class classifier'
)
epoch_acc_2class.plot(
    x='epochs',
    y=['acc_list', 'micro_list', 'macro_list'],
    xlabel='Epochs',
    ylabel='Accuracy',
    grid=True,
    marker='.',
    title='Accuracy, Micro, macro with epochs for 2-class classifier'
)
epoch_acc_10class.plot(
    x='epochs',
    y=['acc_list', 'micro_list', 'macro_list'],
    xlabel='Epochs',
    ylabel='Accuracy',
    grid=True,
    marker='.',
    title='Accuracy, Micro, macro with epochs for 10-class classifier'
)
```

```python
epochs_acc_3 = pd.DataFrame()
epochs_acc_3['acc_2classes'] = epoch_acc_2class['acc_list']
epochs_acc_3['acc_5classes'] = epoch_acc_5class['acc_list']
epochs_acc_3['acc_10classes'] = epoch_acc_10class['acc_list']
epochs_acc_3['epochs'] = epoch_acc_10class['epochs']
print(epochs_acc_3.head())
epochs_acc_3.plot(
    x='epochs',
    y=['acc_2classes', 'acc_5classes', 'acc_10classes'],
    xlabel='Epochs',
    ylabel='Accuracy',
    grid=True,
    marker='.',
    title='Accuracy with epochs for 2, 5, 10 classes'
)
```

```python
epochs_micro_3 = pd.DataFrame()
epochs_micro_3['micro_2classes'] = epoch_acc_2class['micro_list']
epochs_micro_3['micro_5classes'] = epoch_acc_5class['micro_list']
epochs_micro_3['micro_10classes'] = epoch_acc_10class['micro_list']
epochs_micro_3['epochs'] = epoch_acc_10class['epochs']
print(epochs_micro_3.head())
epochs_micro_3.plot(
    x='epochs',
    y=['micro_2classes', 'micro_5classes', 'micro_10classes'],
    xlabel='Epochs',
    ylabel='micro scores',
    grid=True,
    marker='.',
    title='Micro scores with epochs for 2, 5, 10 classes'
)
```

```python
epochs_macro_3 = pd.DataFrame()
epochs_macro_3['macro_2classes'] = epoch_acc_2class['macro_list']
epochs_macro_3['macro_5classes'] = epoch_acc_5class['macro_list']
epochs_macro_3['macro_10classes'] = epoch_acc_10class['macro_list']
epochs_macro_3['epochs'] = epoch_acc_10class['epochs']
print(epochs_macro_3.head())
epochs_macro_3.plot(
    x='epochs',
    y=['macro_2classes', 'macro_5classes', 'macro_10classes'],
    xlabel='Epochs',
    ylabel='macro scores',
    grid=True,
    marker='.',
    title='Macro scores with epochs for 2, 5, 10 classes'
)
```
